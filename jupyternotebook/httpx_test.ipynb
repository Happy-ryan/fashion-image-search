```python
import httpx
```

```python
API_URL = "http://localhost:9000/embedding"
rid = "9da44615-fc5f-49ef-b455-b04367e29845"

async with httpx.AsyncClient() as client:
    response = await client.get(
        f"{API_URL}/{rid}",
        timeout=None
        )
```

```python
print(response)
print(response.json())
embedding = response.json()["embedding"]
print(embedding)
```

```python
API_URL = "http://localhost:9001/search"
async with httpx.AsyncClient() as client:
    response = await client.post(
        f"{API_URL}",
        json={
            "embedding": embedding,
        },
    )

    
print(response)
print(response.status_code)
print(response.json())
```
